```python
#python
p_knn= pd.read_csv("D:/Bachelorarbeit/1.Zahlenerkennung/Python_Ressourcenverbrauch/Python_KNN_process_usage.csv")
p_lr= pd.read_csv("D:/Bachelorarbeit/1.Zahlenerkennung/Python_Ressourcenverbrauch/Python_LR_process_usage.csv")
p_dt= pd.read_csv("D:/Bachelorarbeit/1.Zahlenerkennung/Python_Ressourcenverbrauch/Python_DT_process_usage.csv")
p_rf= pd.read_csv("D:/Bachelorarbeit/1.Zahlenerkennung/Python_Ressourcenverbrauch/Python_RF_process_usage.csv")
p_nb= pd.read_csv("D:/Bachelorarbeit/1.Zahlenerkennung/Python_Ressourcenverbrauch/Python_NB_process_usage.csv")
p_svm= pd.read_csv("D:/Bachelorarbeit/1.Zahlenerkennung/Python_Ressourcenverbrauch/Python_SVM_process_usage.csv")
p_knn['second'] = range(1, len(p_knn) + 1)
p_lr['second'] = range(1, len(p_lr) + 1)
p_dt['second'] = range(1, len(p_dt) + 1)
p_rf['second'] = range(1, len(p_rf) + 1)
p_nb['second'] = range(1, len(p_nb) + 1)
p_svm['second'] = range(1, len(p_svm) + 1)
```

```python
plt.figure(figsize=(15,3))
plt.plot(p_knn['second'],p_knn['Memory usage (MB)'], label = "K Nearest Neighbor")
plt.plot(p_lr['second'],p_lr['Memory usage (MB)'], label = "Logistic Regression")
plt.plot(p_nb['second'],p_nb['Memory usage (MB)'], label = "Naive Bayes")
plt.plot(p_rf['second'],p_rf['Memory usage (MB)'], label = "Random Forest")
plt.plot(p_dt['second'],p_dt['Memory usage (MB)'], label = "Decision Tree")
plt.plot(p_svm['second'],p_svm['Memory usage (MB)'], label = "Support Vector Machine")
plt.xlabel("Ausführungszeit in Sekunden")
plt.ylabel("Arbeitsspeicher in MB")
plt.title("Python")
plt.grid(True)
plt.legend()
```

```python
plt.figure(figsize=(15, 3))
plt.plot(p_knn['second'], p_knn['CPU usage (%)'], label="K Nearest Neighbor")
plt.plot(p_lr['second'], p_lr['CPU usage (%)'], label="Logistic Regression")
plt.plot(p_nb['second'], p_nb['CPU usage (%)'], label="Naive Bayes")
plt.plot(p_rf['second'], p_rf['CPU usage (%)'], label="Random Forest")
plt.plot(p_dt['second'], p_dt['CPU usage (%)'], label="Decision Tree")
plt.plot(p_svm['second'], p_svm['CPU usage (%)'], label="Support Vector Machine")
plt.xlabel("Ausführungszeit in Sekunden")
plt.ylabel("CPU Auslastung in %")
plt.title("Python")
plt.grid(True)
plt.legend()
```
