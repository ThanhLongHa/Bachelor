```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
p = pd.read_csv("D:/Bachelorarbeit/2.Bank_abwanderung/Ressourcenverbrauch/Python_Churn_process_usage.csv")
k = pd.read_csv("D:/Bachelorarbeit/2.Bank_abwanderung/Ressourcenverbrauch/Knime_Churn2_process_usage.csv")
p['second'] = range(1, len(p) + 1)
k['second'] = range(1, len(k) + 1)
```

```python
plt.figure(figsize=(15,3))
plt.plot(p['second'],p['Memory usage (MB)'], label = "Python")
plt.plot(k['second'],k['Memory usage (MB)'], label = "Knime")
plt.legend()
plt.grid(True)
plt.xlabel("Sekunden")
plt.ylabel("RAM Verbrauch in MB")
```

```python
plt.figure(figsize=(15,3))
plt.plot(p['second'],p['CPU usage (%)'], label = "Python")
plt.plot(k['second'],k['CPU usage (%)'], label = "Knime")
plt.legend()
plt.grid(True)
plt.xlabel("Sekunden")
plt.ylabel("CPU Auslastung in %")
```

```python
#Max und Mean
print("Python Max CPU:", p['CPU usage (%)'].max(),"| Mean:",p['CPU usage (%)'].mean())
print("Python Max RAM:", p['Memory usage (MB)'].max(),"| Mean:",p['Memory usage (MB)'].mean())

print("Knime Max CPU:", k['CPU usage (%)'].max(),"| Mean:",k['CPU usage (%)'].mean())
print("Knime Max RAM:", k['Memory usage (MB)'].max(),"| Mean:",k['Memory usage (MB)'].mean())
```
